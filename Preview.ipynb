```python
from collections import Counter
from sklearn.datasets import make_classification
from imblearn.over_sampling import SMOTE 
```

```python
sm = SMOTE(random_state = 42)
Y = df.iloc[:, -1].values
X = df.iloc[:,0:77].values
X
X_train_oversampled, y_train_oversampled = sm.fit_resample(X, Y)
```

```python
w=Counter(y_train_oversampled)
w
```

```python
import matplotlib.pyplot as plt
plt.bar(w.keys(), w.values())
```

```python
from sklearn.model_selection import train_test_split

X_Train, X_Test, Y_Train, Y_Test = train_test_split(X_train_oversampled, y_train_oversampled, test_size = 0.25, random_state = 0)
```

Random Forest

```python
X_train_oversampled
```

```python
from sklearn.preprocessing import StandardScaler
sc_X = StandardScaler()
X_Train = sc_X.fit_transform(X_Train)
X_Test = sc_X.transform(X_Test)
```

```python
from sklearn.ensemble import RandomForestClassifier
classifier = RandomForestClassifier(n_estimators = 50, criterion = 'entropy', random_state = 0)
classifier.fit(X_Train,Y_Train)
```

```python
Y_Pred = classifier.predict(X_Test)


from sklearn.metrics import confusion_matrix
cm = confusion_matrix(Y_Test, Y_Pred)
print(cm)
```

```python
import matplotlib.pyplot as plt
import seaborn as sn

conf_matrix=pd.DataFrame(data=cm,columns=['Predicted:0','Predicted:1'],index=['Actual:0','Actual:1'])
plt.figure(figsize = (8,5))
sn.heatmap(conf_matrix, annot=True,fmt='d')
```

```python
#Accuracy
import sklearn
rf_acc=sklearn.metrics.accuracy_score(Y_Test, Y_Pred)
```
